```python
import pandas as pd
temperature = pd.read_csv("temperature.csv")
temperature.head()
temperature.info
```

```python
#Dotaz na měření, která byla provedena v Praze. Je na datech něco zvláštního? Napadá tě, čím to může být? Zde je nápověda.
praha = temperature[temperature["City"] == "Prague"]
print(praha[["City", "Day", "AvgTemperature"]])
#data jsou uvedena ve stupnici Fahrenheita
```

```python
#Dotaz na měření, ve kterých je teplota (sloupec AvgTemperature) vyšší než 80 stupňů.

teplota = temperature[temperature["AvgTemperature"] >= 80]
print(teplota[["Region", "Country", "City", "Day", "AvgTemperature"]])
```

```python
#Dotaz na měření, ve kterých je teplota vyšší než 60 stupňů a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).


teplota2 = temperature[(temperature["AvgTemperature"] >= 60) & (temperature["Region"] == "Europe")]
print(teplota2[["Region", "Country", "City", "Day", "AvgTemperature"]])
```

```python
#Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 80 stupňů nebo menší než -20 stupňů.

teplota3 = temperature[(temperature["AvgTemperature"] >= 60) | (temperature["AvgTemperature"] <= -20)]
print(teplota3[["Region", "Country", "City", "Day", "AvgTemperature"]])
teplota3.sort_values(by="AvgTemperature")
```
