```python
what_is_this = adopce[adopce["id"] == 35]
print(what_is_this[["id", "nazev_cz", "nazev_en"]])
#důkaz, že kod funguje
```

```python
what_is_this = adopce[adopce["id"] == 34]
print(what_is_this[["id", "nazev_cz", "nazev_en"]])
#vypadá to, že kod funguje správně, ale pod ID 34 není žádné zvíře
```

```python
#nepovinny ukol

adopce.sort_values(by="id")
```

```python
#Adopce kterých zvířat je nejdražší?
#Nejdražší jsou Tygr ussurijský, Tygr sumaterský, Slon indický, Hroch obojživelný, Žirafa severní núbijský a Tygr malajský
adopce.sort_values(by="cena", ascending=False)
nejdražší = adopce[adopce["cena"] >= 50000]
print(nejdražší[["id", "nazev_cz", "cena"]])
```

```python
#Které zvíře je alfabeticky poslední v češtině? Které v angličtině?
#CZ - Žluva hajní
adopce.sort_values(by="nazev_cz", ascending=False)
```

```python
#Které zvíře je alfabeticky poslední v češtině? Které v angličtině?
#EN - Yellow-necked francolin
adopce.sort_values(by="nazev_en", ascending=False)
```
