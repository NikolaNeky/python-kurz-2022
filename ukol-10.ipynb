```python
import pandas as pd
zam_praha = pd.read_csv("zam_praha.csv")
zam_plzen = pd.read_csv("zam_plzeň.csv")
zam_liberec = pd.read_csv("zam_liberec.csv")

zam_praha.head()
zam_plzen.head()
zam_liberec.head()

zam_praha.info
zam_plzen.info
zam_liberec.info
```

```python
zam_praha["mesto"] = "Praha"
zam_plzen["mesto"] = "Plzeň"
zam_liberec["mesto"] = "Liberec"


zam_plzen.head()
```

```python
zamestnanci = pd.concat([zam_praha, zam_plzen, zam_liberec], ignore_index = True)

zamestnanci.info
```

```python
import pandas as pd
platy_2021_02 = pd.read_csv("platy_2021_02.csv")

platy_2021_02.info
```

```python
merge_all = pd.merge(zamestnanci, platy_2021_02, on=["cislo_zamestnance"])

merge_all.info
```

```python
#průměrný plat je

print(merge_all.groupby('mesto')["plat"].mean())
```

```python
#Zadani 2

import pandas as pd
vykazy = pd.read_csv("vykazy.csv")

vykazy.head
```

```python
#prejmenovat sloupec

vykazy.rename(columns={"emloyee_id": "cislo_zamestnance"}, inplace=True)
vykazy.head
```

```python
print(vykazy['cislo_zamestnance'].unique())
print(merge_all['cislo_zamestnance'].unique())
```

```python
#zmergovat na základě čísla zaměstnance

Merge_pro_vykaz = pd.merge(vykazy, merge_all, on=["cislo_zamestnance"], how="outer")
Merge_pro_vykaz.info
```

```python
# zjisti celkový počet vykázaných hodin za jednotlivé projekty.

Celkovy_pocet_hodin = Merge_pro_vykaz.groupby("project")["hours"].sum()

Celkovy_pocet_hodin.info
```
